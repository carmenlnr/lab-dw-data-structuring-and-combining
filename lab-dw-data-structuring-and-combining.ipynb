## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

df1.columns = df1.columns.str.lower().str.strip().str.replace(" ", "_")
df2.columns = df2.columns.str.lower().str.strip().str.replace(" ", "_")
df3.columns = df3.columns.str.lower().str.strip().str.replace(" ", "_")

# unimos (concatenar filas)
df = pd.concat([df1, df2, df3], ignore_index=True)

# vemos estructura inicial
df.head()
```

```python
df[["state", "st"]].isna().mean()
```

```python
df[["state", "st"]].nunique()
```

```python
df["state"] = df["state"].fillna(df["st"])
df = df.drop(columns=["st"])
df.head()
```

```python
#limpiamos customer lifetime value
df["customer_lifetime_value"] = (
    df["customer_lifetime_value"]
    .astype(str)
    .str.replace("%", "")
)

df["customer_lifetime_value"] = pd.to_numeric(df["customer_lifetime_value"], errors="coerce")
```

```python
#limpiamos income y otros
df["income"] = pd.to_numeric(df["income"], errors="coerce")
df["monthly_premium_auto"] = pd.to_numeric(df["monthly_premium_auto"], errors="coerce")
df["total_claim_amount"] = pd.to_numeric(df["total_claim_amount"], errors="coerce")
```

```python
#limpiamos texto
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip().str.lower()
```

```python
df.info()
df.head()
```

```python
df["number_of_open_complaints"].value_counts().head()
df["number_of_open_complaints"].unique()[:10]
```

```python
# falta limpiar complaints

df["number_of_open_complaints"] = (
    df["number_of_open_complaints"]
    .astype(str)
    .str.split("/").str[1]
)

df["number_of_open_complaints"] = pd.to_numeric(df["number_of_open_complaints"], errors="coerce")
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
channels = ["branch", "call_center", "web", "mail"]

df_filtered = df[df["sales_channel"].isin(channels)]
```

```python
revenue_pivot = df.pivot_table(
    index="sales_channel",
    values="total_claim_amount",
    aggfunc="sum"
).round(2).sort_values(by="total_claim_amount", ascending=False)

revenue_pivot
```

El canal de Agentes (Agent) es el más efectivo, generando 1,810,226.82 en ingresos totales. Este resultado indica que la interacción directa con agentes es el principal motor de ventas y generación de valor para la empresa.

El canal de Sucursal (Branch) ocupa el segundo lugar con 1,301,204.00, mostrando un rendimiento sólido, aunque significativamente inferior al canal de agentes.

El Call Center genera 926,600.82, lo que representa una contribución intermedia. Aunque sigue siendo relevante, su eficiencia es menor en comparación con los canales presenciales.

Por último, el canal Web es el que menor ingresos genera con 706,600.04, lo que sugiere un bajo rendimiento del canal digital en comparación con los canales tradicionales.

- nota : El enunciado menciona el canal “mail”, pero este no aparece en el dataset. En su lugar, el dataset incluye el canal “Agent”, por lo que el análisis se ha realizado utilizando las categorías reales disponibles en los datos.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
df["customer_lifetime_value"].dtype
```

```python
df["customer_lifetime_value"] = pd.to_numeric(
    df["customer_lifetime_value"],
    errors="coerce"
)
```

```python
clv_pivot = df.pivot_table(
    index=["gender", "education"],
    values="customer_lifetime_value",
    aggfunc="mean"
).round(2)

clv_pivot.unstack()
```

El Customer Lifetime Value (CLV) varía ligeramente entre segmentos de género y nivel educativo, aunque no se observan diferencias extremadamente grandes.

En el caso de las mujeres (F), el mayor CLV se encuentra en el nivel educativo de High School or Below (8675.22), seguido de Master (8157.05). En contraste, el valor más bajo aparece en el nivel de Doctor (7328.51).

Para los hombres (M), los valores de CLV son relativamente estables, destacando ligeramente Master (8168.83) y College (8052.46) como los segmentos más altos. El valor más bajo se observa en Bachelor (7703.60).

Las diferencias en CLV entre los segmentos son relativamente pequeñas, lo que sugiere que no hay una segmentación fuerte por género o nivel educativo. Por tanto, genero y nivel educativo parece ser un factor determinante fuerte. La empresa podría obtener mejores resultados enfocándose en otros factores como el canal de venta o el tipo de póliza.
